```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

print(f"Shares skewness: {skew(data['shares'])}")
sns.histplot(data['shares'], kde=True, log_scale=True)
plt.savefig('histogram.png')
plt.show()
sns.boxplot(data['shares'], orient='h', log_scale=True, whis=3)
```

```python
sorted_shares = data.sort_values('shares') 
median = sorted_shares['shares'].median() 
q1 = sorted_shares['shares'].quantile(q=0.25) 
q3 = sorted_shares['shares'].quantile(q=0.75) 
iqr = q3 - q1
```

```python
lower_fence = q1 - (iqr * 1.5) 
upper_fence = q3 + (iqr * 1.5)
print(lower_fence)
print(upper_fence)
```

```python
data2 = data[(data['shares'] <= upper_fence) & (data['shares'] >= lower_fence)]
print(data2.shape)
# data3 = data2[data2['shares'] >= lower_fence]
# print(data3.shape)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split

#oddzielenie danych i targetw
dane=data2.iloc[:,0:58]
dane=dane.to_numpy()
target=data2.iloc[:,-1]
target=target.to_numpy()

#podzia danych na zestaw treningowy i testowy
X_train, X_test, y_train, y_test = train_test_split(dane, target, train_size=0.9, shuffle=True)

#zamiana tablic na tensory
x = torch.tensor(X_train, dtype=torch.float32)
targets = torch.tensor(y_train, dtype=torch.float32).reshape(-1, 1)

x_test = torch.tensor(X_test, dtype=torch.float32)
target_test = torch.tensor(y_test, dtype=torch.float32).reshape(-1, 1)
```

```python
import torch
from torchmetrics.regression import R2Score
from matplotlib import pyplot as plt

#funkcja do trenowania, testowania i ewaluacji modeli sieci neuronowych 

def trenTest(model, b_size=1000, decay=1e-6):
    
    #utworzenie dataset i dataloader
    dataset = TensorDataset(x,targets)
    dataloader = DataLoader(dataset, batch_size=b_size, shuffle=True)

    #tabela na wyniki funkcji straty
    epoch_loss=[]

    optimizer = optim.Adam(model.parameters(), weight_decay=decay)

    w=[0,99,199,299,399,499,599,699,799,899,999]
    
    for epoch in range(1000):
        #train model, ustawienie etykiety e model jest trenowany
        model.train()
        for data in dataloader:
            #zerowanie gradientw
            optimizer.zero_grad()
            a,b = data
            model_estimation = model(a)
            #print(model_estimation)
            loss = mse(model_estimation, b)
            loss.backward() #calculate gradients in neural network using backpropagation alg
            #break
            optimizer.step() #update weights in neural network
    
        #evaluate model at every epoch
        model.eval()
        y_pred = model(x_test)
        loss = mse(y_pred, target_test)
    
        if epoch in w:
            epoch_loss.append(loss.detach().numpy())

    #wykres wynikw f. straty na co 100 epok
    plt.plot(w,epoch_loss)
    plt.title("Wykres wartoci f. straty do liczby epok")
    plt.xlabel('epoki')
    plt.ylabel('loss')
    plt.show()

    #ewaluacja modelu
    id=9
    prediction=model(x_test[id])
    
    print("Model predicted for article {0} value {1}".format(id, prediction))
    print("Real value for article \"{0}\" is {1}".format(id, target_test[id]))
    
    r2=R2Score()
    print("R2:")
    print(r2(model(x_test),target_test))
```
